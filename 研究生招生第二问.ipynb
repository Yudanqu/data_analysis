```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 为了防止画图时乱码
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
gaoxiao_df = pd.read_excel('buchong.xls')
gaoxiao_df.columns = list(gaoxiao_df.iloc[0])
gaoxiao_df = gaoxiao_df[1:]
```

```python
gaoxiao_df.info()
```

```python
gaoxiao_df.head(3)
```

```python
df_person_num = gaoxiao_df.groupby('2007-2011年的岗位等级')['硕士招生人数2007','硕士招生人数2008','硕士招生人数2009','硕士招生人数2010','硕士招生人数2011'].sum()
df_person_num
```

```python
df_person_num.index
```

```python
index = ['一级岗', '二级岗', '三级岗', '四级岗', '五级岗', '六级岗', '七级岗']
df_person_num.reindex(index) # 索引重排序
df_person_num.plot()
plt.title('硕士招生人数')
plt.savefig("硕士招生人数.jpg")
plt.show()
```

```python
df_english_num = gaoxiao_df.groupby('2007-2011年的岗位等级')['发表英文论文篇数2007','发表英文论文篇数2008','发表英文论文篇数2009','发表英文论文篇数2010','发表英文论文篇数2011'].sum()
df_english_num.reindex(index)
df_english_num.plot()
plt.title('发表英文论文篇数')
plt.savefig("发表英文论文篇数.jpg")
plt.show()
```

```python
df_chinese_num = gaoxiao_df.groupby('2007-2011年的岗位等级')['中文期刊论文篇数2007','中文期刊论文篇数2008','中文期刊论文篇数2009','中文期刊论文篇数2010','中文期刊论文篇数2011'].sum()
df_chinese_num.reindex(index)
df_chinese_num.plot()
plt.title('中文期刊论文篇数')
plt.savefig("中文期刊论文篇数.jpg")
plt.show()
```

```python
df_patent_num = gaoxiao_df.groupby('2007-2011年的岗位等级')['申请专利数2007','申请专利数2008','申请专利数2009','申请专利数2010','申请专利数2011'].sum()
df_patent_num.reindex(index)
df_patent_num.plot()
plt.title('申请专利数')
plt.savefig("申请专利数.jpg")
plt.show()
```

```python
df_award_num = gaoxiao_df.groupby('2007-2011年的岗位等级')['获得奖励个数2007','获得奖励个数2008','获得奖励个数2009','获得奖励个数2010','获得奖励个数2011'].sum()
df_award_num.reindex(index)
df_award_num.plot()
plt.title('获得奖励个数')
plt.savefig("获得奖励个数.jpg")
plt.show()
```

# 平均值

```python
amount1 = gaoxiao_df.groupby('2007-2011年的岗位等级')['硕士招生人数2007','硕士招生人数2008','硕士招生人数2009','硕士招生人数2010','硕士招生人数2011'].count()
new_df_person_num = df_person_num / amount1
new_df_person_num.reindex(index) # 索引重排序
new_df_person_num
new_df_person_num.plot()
plt.title('硕士招生人数')
plt.savefig("硕士招生人数.jpg")
plt.show()
```

```python
df_paper_num_s = gaoxiao_df.groupby('2007-2011年的岗位等级')['获得省优硕论文篇数'].sum()
df_paper_num_x = gaoxiao_df.groupby('2007-2011年的岗位等级')['获得校优硕论文篇数'].sum()
df_paper_num_s = df_paper_num_s.reindex(index) # 省优秀论文
df_paper_num_x = df_paper_num_x.reindex(index) # 校优秀论文
df_all = df_paper_num_s + df_paper_num_x # 总优秀论文
```

```python
fig=plt.figure()
df_paper_num_s.plot(kind='bar', fc='r')
df_paper_num_x.plot(kind='bar', bottom=df_paper_num_s, fc='b')
plt.title('各级岗获优秀论文数量')
plt.legend(['获得省优硕论文篇数', '获得校优硕论文篇数'])
plt.savefig('获奖数量.jpg')
plt.show()
```
