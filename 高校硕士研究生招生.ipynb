计算平均每级岗位对应的个因素的平均值

```python
df_qiuhe = gaoxiao_df.groupby('2007-2011年的岗位等级').sum()
```

```python
df1 = df_qiuhe.copy()
columns = ['硕士招生人数合计', '到账经费5年合计', '发表英文论文篇数合计', '发表英文论文篇数影响因子篇均值', '中文期刊论文篇数合计', '申请专利数合计', '获得奖励个数合计']
new_df = pd.DataFrame(df1, columns=columns)
print(new_df.head(5))
df2 = ave_gangwei.copy()
new_ave_gangwei = pd.DataFrame(df2, columns=columns)
print(new_ave_gangwei)
```
